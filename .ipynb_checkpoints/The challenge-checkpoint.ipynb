```python
genres = 'blues classical country disco hiphop jazz metal pop reggae rock'.split()
```

```python
features = []

columns = ['Song_id', 'chroma_stft', 'rmse', 'spec_cent', 'spec_bw', 'rolloff', 'zcr', 'mfcc']

for g in genres:

    for filename in os.listdir(f'./drive/My Drive/Colab Notebooks/genres/{g}'):

        songname = f'./drive/My Drive/Colab Notebooks/genres/{g}/{filename}'
        y, sr = librosa.load(songname, mono=True, duration=30)
        rmse = librosa.feature.rmse(y=y)
        chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
        spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
        spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
        rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
        zcr = librosa.feature.zero_crossing_rate(y)
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
        to_append = np.array([filename, np.mean(chroma_stft), np.mean(rmse), np.mean(spec_cent), np.mean(spec_bw), np.mean(rolloff), np.mean(zcr), np.mean(mfcc)])    
        
        #for e in mfcc:
          #  to_append += f' {np.mean(e)}'
        #to_append += f' {g}'
        features.append(to_append)

featuresdf = pd.DataFrame(features, columns=columns)

featuresdf.head()
```

```python
featuresdf['Rating'] = np.random.randint(90.0, 100.0, featuresdf.shape[0])

featuresdf.head()
```

```python
X = np.array(featuresdf[['Song_id','chroma_stft', 'spec_cent', 'spec_bw', 'rolloff', 'zcr', 'mfcc']])

y = np.array(featuresdf['Rating'],  dtype=np.float)

encoder = LabelEncoder()

featuresdf['chroma_stft'] = encoder.fit_transform(featuresdf['chroma_stft'])
featuresdf['spec_cent'] = encoder.fit_transform(featuresdf['spec_cent'])
featuresdf['spec_bw'] = encoder.fit_transform(featuresdf['spec_bw'])
featuresdf['rolloff'] = encoder.fit_transform(featuresdf['rolloff'])
featuresdf['zcr'] = encoder.fit_transform(featuresdf['zcr'])
featuresdf['mcff'] = encoder.fit_transform(featuresdf['mfcc'])

yy = to_categorical(encoder.fit_transform(y))
```

```python
featuresdf.head()
#print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=45)

print(X_train.shape, y_train.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Convolution2D, Conv2D, MaxPooling2D, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from sklearn import metrics 
from tensorflow.keras.callbacks import ModelCheckpoint 
```

```python
print(X_test.shape, y_test.shape)

print(type(X_train))
```

```python
num_rows = 7
num_columns = 750
num_channels = 1


X_train = X_train.reshape(num_rows, num_columns, num_channels)
#X_test = X_test.reshape(7, X_test.shape[0], num_channels)

num_labels = yy.shape[1]

model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(1,1), input_shape=( num_rows, num_columns, num_channels) ,activation='relu'))
model.add(MaxPooling2D(pool_size=(1,1)))
model.add(Dropout(0.2))

model.add(Conv2D(filters=32, kernel_size=2, activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.2))

model.add(Conv2D(filters=64, kernel_size=2, activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.2))

model.add(Conv2D(filters=128, kernel_size=2, activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.2))
model.add(GlobalAveragePooling2D())
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')

# Display model architecture summary 
model.summary()



# Calculate pre-training accuracy 
#score = model.evaluate(X_test, y_test, verbose=1)
#accuracy = 100*score[1]

#print("Pre-training accuracy: %.4f%%" % accuracy)
```

```python
num_epochs = 72
num_batch_size = 256

checkpointer = ModelCheckpoint(filepath='saved_models/weights.best.basic_cnn.hdf5', verbose=1, save_best_only=True)

model.fit(X_train, y_train, batch_size=num_batch_size, epochs=num_epochs,  callbacks=[checkpointer], verbose=1)
```
